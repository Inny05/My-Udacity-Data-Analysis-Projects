## Investigation Overview

>The Ford Gobike dataset contains information about individual rides made in a bike sharing company. This dataset originally contained 183412 rows and 16 columns, and after cleaning we were left with 174952 columns and 23 rows as some rows were dropped and new columns were created.

> Over 70 percent of the users are males.
> 90.5% of users are subscribers and 9.5% are customers.
> The highest number of rides were recorded on Thurday and the least number of rides were recorded on Saturday and Sunday, 8.2% and 8.3% respectively.
> The most number of bike usage was recorded at 8am-9am and 5pm-6pm. 


## Dataset Overview

> I cleaned the Ford Gobike dataset by dropping columns that contianed missing values. I also converting the columns with time values from integers to datetime datatypes and I converted columns like user type, bike sharing and member gender to categorical datatypes.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df_bike = pd.read_csv('df_bike.csv')
````

## (Visualization 1)

> This visualization shows the number of daily rides for the week, with the most rides recorded on Thursdays and the least recorded on Saturdays and Sundays.  

````python
base_color = sb.color_palette()[0]
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday','Saturday', 'Sunday']
plt.figure(figsize =[15,6])
sb.countplot(data= df_bike , x='start_day' , color= base_color, order= days_order)
plt.title('Bike ride most popular days', fontsize = 16, fontweight='bold')
plt.ylabel('Number of Bike trips', fontsize = 15)
plt.xlabel('Bike Ride start days', fontsize= 15)
plt.title('Daily Bike Rides', fontsize=14, fontweight='bold')
plt.xlabel('Days');
days_count = df_bike['start_day'].value_counts()
total_trips= days_count.sum()
plt.yticks(size=13); 
locs, labels = plt.xticks( size=14); 
for loc, label in zip(locs, labels):
    count = days_count[label.get_text()]
    string = '{:0.1f}%'.format(100*count/total_trips)
    plt.text(loc, count-10000, string, ha = 'center', color = 'white',fontsize = 13)
````

## (Visualization 2)

> This visualisations show the relationships between the user type and the number of trips, number of trips and bike sharing for all trips and the number of trips by the gender types.

````python
fig, ax = plt.subplots(3,figsize = [10,10])
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday','Saturday', 'Sunday']
sb.countplot(data= df_bike, x = 'start_day', hue = 'user_type',order= days_order,ax=ax[0])
sb.countplot(data= df_bike, x = 'start_day', hue = 'bike_share_for_all_trip', order= days_order,ax=ax[1])
sb.countplot(data= df_bike, x = 'start_day', hue = 'member_gender', order= days_order, ax=ax[2])
````

## (Visualization 3)

> This plot shows the relationship between gender types, bike sharing and the average duration of trips. More females are prone to take longer trips bike sharing while, few males take longer trips without bike sharing. The other gender

````python
fig = plt.figure(figsize = [8,6])
base_color = sb.color_palette()[0]
ax = sb.pointplot(data = df_bike, x ='member_gender', y = 'duration_sec', hue = 'bike_share_for_all_trip', 
            color= base_color, linestyles = '', dodge = 0.4)
plt.title('Trip Duration across gender and bike_share_for_all_trip', fontsize=14, fontweight='bold')
plt.ylabel('Average Trip Duration (Secs)')
plt.xlabel('Gender', fontsize = 14)
ax.set_yticklabels([],minor = True)
plt.show()
````
